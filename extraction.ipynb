```python
import PyPDF2

def extract_text_from_pdf(pdf_path):
    text = ""
    try:
        with open(pdf_path, "rb") as f:
            reader = PyPDF2.PdfReader(f)
            num_pages = len(reader.pages)
            for page_num in range(num_pages):
                page =reader.pages[page_num]
                text += page.extract_text()
    except Exception as e:
        print(f"Error extracting text from PDF: {e}")

    return text

# Example usage:
pdf_path = "/content/CN Mod1@AzDOCUMENTS.in.pdf"  # Replace this with the path to your PDF file
extracted_text = extract_text_from_pdf(pdf_path)
print(extracted_text)


def divide_text_into_chunks(text, chunk_size=500):
    """
    Divide the extracted text into chunks based on a specified size.
    """
    chunks = []
    words = text.split()
    current_chunk = ""

    for word in words:
        if len(current_chunk) + len(word) < chunk_size:
            current_chunk += word + " "
        else:
            chunks.append(current_chunk)
            current_chunk = word + " "

    # Append the last chunk
    if current_chunk:
        chunks.append(current_chunk)

    return chunks


chunks = divide_text_into_chunks(extract_text_from_pdf(pdf_path))
for i, chunk in enumerate(chunks):
    print(f"Chunk {i+1}: {chunk}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Define a function to calculate cosine similarity between user input and chunks
def calculate_similarity(user_input, chunks):
    # Add user input to the chunks
    chunks.append(user_input)

    # Convert chunks to TF-IDF vectors
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(chunks)

    # Calculate cosine similarity matrix
    similarity_matrix = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1])

    return similarity_matrix

# Example usage:
user_input = input("Enter your query: ")  # Prompt user for input

# Calculate similarity
similarity_matrix = calculate_similarity(user_input, chunks)

# Print similarity scores for each chunk
for i, similarity_score in enumerate(similarity_matrix[0]):
    print(f"Similarity score with Chunk {i+1}: {similarity_score}")
```

```python
import re

def divide_text_into_paragraphs(text):
    """
    Divide the extracted text into paragraphs.
    """
    # Split text into paragraphs using regular expression
    paragraphs = re.split(r'(?<=\w\.)\s+(?=[A-Z])', text)

    return paragraphs

pdf_path = "/content/CN Mod1@AzDOCUMENTS.in.pdf"
extracted_text = extract_text_from_pdf(pdf_path)
paragraphs = divide_text_into_paragraphs(extracted_text)
for i, paragraph in enumerate(paragraphs):
    print(f"Paragraph {i+1}: {paragraph}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def calculate_similarity(user_input, paragraphs):
    """
    Calculate the similarity between user input and each paragraph.
    """
    # Add user input to the list of paragraphs
    paragraphs.append(user_input)

    # Convert paragraphs to TF-IDF vectors
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(paragraphs)

    # Calculate cosine similarity between user input and paragraphs
    similarity_scores = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1]).flatten()

    # Remove the user input from the list of paragraphs
    paragraphs.pop()

    return similarity_scores

# Example usage:
user_input = input("Enter your query: ")  # Prompt user for input

# Calculate similarity scores
similarity_scores = calculate_similarity(user_input, paragraphs)

# Print similarity scores for each paragraph
for i, score in enumerate(similarity_scores):
    print(f"Similarity score with Paragraph {i+1}: {score}")
```

```python
def calculate_similarity(user_input, paragraphs):
    """
    Calculate the similarity between user input and each paragraph.
    Return the most similar 5 paragraphs.
    """
    # Convert paragraphs to TF-IDF vectors
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(paragraphs + [user_input])

    # Calculate cosine similarity between user input and paragraphs
    similarity_scores = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1]).flatten()

    # Get indices of top 5 most similar paragraphs
    top_indices = similarity_scores.argsort()[-5:][::-1]

    # Retrieve the most similar 5 paragraphs
    most_similar_paragraphs = [paragraphs[i] for i in top_indices]

    return most_similar_paragraphs

# Example usage:
user_input = input("Enter your query: ")  # Prompt user for input

# Get the most similar 5 paragraphs
most_similar_paragraphs = calculate_similarity(user_input, paragraphs)

# Print the most similar 5 paragraphs
print("Most similar 5 paragraphs:")
for i, paragraph in enumerate(most_similar_paragraphs):
    print(f"Paragraph {i+1}: {paragraph}")
```
